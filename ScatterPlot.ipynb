```python
%matplotlib inline


import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

charity_data = "cleaned_data.csv"
charity_df = pd.read_csv("cleaned_data.csv")
charity_df.head()
```

```python
charity_x_axis = charity_df["Contributions"]

charity_y_axis = charity_df["Rating"]

plt.scatter(charity_x_axis, charity_y_axis, marker="o", alpha= 0.8,
            color="red", edgecolors="black", linewidths=1, label="Charities")


plt.title("Charity Ratings vs Total Contributions")
plt.xlabel("Annual Contributions")
plt.ylabel("Charity Rating")
plt.xlim(0,10000000)
plt.show
```
